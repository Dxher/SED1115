```python
my_list = [1,3,6,8,50,56,57,78,89,123,124,390,456,567,888,999,1000]
search_value = 123

def binary_search(list, search_value: int):
    list_length = len(list)
    half_list = list_length//2
    if list[half_list] == search_value:
        return "Item is found at: {half_list}"
    else:
        if half_list == 0:
            return "Item not found"
        if list[half_list] > search_value:

            return binary_search(list[half_list:], search_value)
            
        elif list[half_list] < search_value:
            return binary_search(list[:half_list], search_value)
        

answer = binary_search(my_list, search_value)

print(answer)

```
